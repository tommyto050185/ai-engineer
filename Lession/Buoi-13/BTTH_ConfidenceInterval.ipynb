### Kho·∫£ng tin c·∫≠y - Confidence Interval (CI)

```python
import numpy as np
import pandas as pd
import os

# Construct the path to the data file in an OS-independent way
data_dir='data'
filename='satisfaction_survey.csv'
# Get the current working directory of the script
current_dir = os.getcwd()

# Construct the path to the data file by going up one level ('..')
# and then into the 'data' folder.
file_path = os.path.join(current_dir, '..', '..', data_dir, filename)

# Normalize the path to resolve '..' and get the absolute path
file_path = os.path.abspath(file_path)

df = pd.read_csv(file_path)
display(df)
```

```python
### T√≠nh trung b√¨nh m·∫´u ###
avg_sample_pd = df['SatisfactionScore'].mean()
display(avg_sample_pd)
avg_sample_np = np.mean(df['SatisfactionScore'])
print("Trung b√¨nh m·∫´u c·ªßa satisfaction_score:", avg_sample_np)
```

```python
### T√≠nh ƒê·ªô l·ªách chu·∫©n ###
std_sample_pd = df['SatisfactionScore'].std()
display(std_sample_pd)  
```

```python
### Tinh sai s·ªë chu·∫©n ###
n = len(df['SatisfactionScore'])
std_error = std_sample_pd / np.sqrt(n)
print("Sai s·ªë chu·∫©n c·ªßa SatisfactionScore:", std_error)
```

D∆∞·ªõi ƒë√¢y l√† b·∫£ng t-score (t-critical) cho b·∫≠c t·ª± do ùëëùëì = n-1 = 100-1 = 99
df=99 v·ªõi c√°c m·ª©c tin c·∫≠y th∆∞·ªùng d√πng (hai ph√≠a):
M·ª©c tin c·∫≠y (%)	t-critical (df=99)
90%	1.6604
95%	1.9842
98%	2.3646
99%	2.6264

```python
### Sai s·ªë bi√™n ƒë·ªô ###
t_critical = 1.9842  # t·ª´ b·∫£ng t v·ªõi df=99 v√† m·ª©c tin c·∫≠y 95%
margin_of_error = t_critical * std_error
print("Sai s·ªë bi√™n ƒë·ªô c·ªßa SatisfactionScore:", margin_of_error)
```

```python
### kho·∫£ng tin c·∫≠y ###
confidence_interval_lower = avg_sample_np - margin_of_error
confidence_interval_upper = avg_sample_np + margin_of_error
print("Kho·∫£ng tin c·∫≠y 95% c·ªßa SatisfactionScore: ({}, {})".format(confidence_interval_lower, confidence_interval_upper))
```
