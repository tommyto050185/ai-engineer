Vi·∫øt thu·∫ßn v√† g·∫ßn nh∆∞ kh√¥ng x√†i th∆∞ vi·ªán ƒë·ªÉ ki·ªÉm ƒë·ªãnh

```python
import numpy as np
# L·∫•y d·ªØ li·ªáu l∆∞∆°ng
salaries = df['Dataset']
n = len(salaries)
sample_mean = np.mean(salaries)
sample_std = np.std(salaries, ddof=1)
print(sample_mean)
print(sample_std)
```

ƒê·∫∑t gi·∫£ thi·∫øt:
H0: L∆∞∆°ng trung b√¨nh = 113000
H1: L∆∞∆°ng trung b√¨nh kh√°c 113000

```python
from scipy import stats
# Th√¥ng tin trung b√¨nh gi·∫£ ƒë·ªãnh
mu = 113000

# Ki·ªÉm tra gi·∫£ thi·∫øt khi bi·∫øt trung b√¨nh v√† ƒë·ªô l·ªách chu·∫©n t·ªïng th·ªÉ
sigma = 15000
# T√≠nh z_stat l√† v·ªã tr√≠ c·ªßa sample_mean so v·ªõi mu tr√™n tr·ª•c ngang c·ªßa ph√¢n ph·ªëi chu·∫©n
z_stat = (sample_mean - mu)/(sigma/np.sqrt(n))
z_p_two_tailed = 2*(1 - stats.norm.cdf(abs(z_stat)))
```

```python
alpha = 0.05
# ========== In k·∫øt qu·∫£ ==========
print("=== Th·ªëng k√™ m·∫´u ===")
print(f"S·ªë l∆∞·ª£ng m·∫´u (n): {n}")
print(f"Trung b√¨nh m·∫´u: {sample_mean:.2f}")
print(f"ƒê·ªô l·ªách chu·∫©n m·∫´u: {sample_std:.2f}\n")

print("=== Ki·ªÉm ƒë·ªãnh Z (bi·∫øt œÉ = 15000) ===")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value (2 ph√≠a): {z_p_two_tailed:.6f}")

if z_p_two_tailed < alpha:
  print("Lo·∫°i b·ªè H0: l∆∞∆°ng trung b√¨nh kh√°c 113000")
else:
  print("H0 ƒë√∫ng: l∆∞∆°ng trung b√¨nh = 113000")
```

ƒê·∫∑t gi·∫£ thi·∫øt: H0: L∆∞∆°ng trung b√¨nh <= 113000 H1: L∆∞∆°ng trung b√¨nh > 113000

```python
z_p_right = 1 - stats.norm.cdf(z_stat)
alpha = 0.05
# ========== In k·∫øt qu·∫£ ==========
print("=== Th·ªëng k√™ m·∫´u ===")
print(f"S·ªë l∆∞·ª£ng m·∫´u (n): {n}")
print(f"Trung b√¨nh m·∫´u: {sample_mean:.2f}")
print(f"ƒê·ªô l·ªách chu·∫©n m·∫´u: {sample_std:.2f}\n")

print("=== Ki·ªÉm ƒë·ªãnh Z (bi·∫øt œÉ = 15000) ===")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value (2 ph√≠a): {z_p_right:.6f}")

if z_p_right < alpha:
  print("Lo·∫°i b·ªè H0: l∆∞∆°ng trung b√¨nh kh√°c 113000")
else:
  print("H0 ƒë√∫ng: l∆∞∆°ng trung b√¨nh <= 113000")
```

ƒê·∫∑t gi·∫£ thi·∫øt: H0: L∆∞∆°ng trung b√¨nh >= 113000 H1: L∆∞∆°ng trung b√¨nh < 113000

```python
z_p_left = stats.norm.cdf(z_stat)
alpha = 0.05
# ========== In k·∫øt qu·∫£ ==========
print("=== Th·ªëng k√™ m·∫´u ===")
print(f"S·ªë l∆∞·ª£ng m·∫´u (n): {n}")
print(f"Trung b√¨nh m·∫´u: {sample_mean:.2f}")
print(f"ƒê·ªô l·ªách chu·∫©n m·∫´u: {sample_std:.2f}\n")

print("=== Ki·ªÉm ƒë·ªãnh Z (bi·∫øt œÉ = 15000) ===")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value (2 ph√≠a): {z_p_left:.6f}")

if z_p_left < alpha:
  print("Lo·∫°i b·ªè H0: l∆∞∆°ng trung b√¨nh < 113000")
else:
  print("H0 ƒë√∫ng: l∆∞∆°ng trung b√¨nh >= 113000")
```

Ki·ªÉm ƒë·ªãnh b·∫±ng th∆∞ vi·ªán c√≥ s·∫µn

```python
from statsmodels.stats.weightstats import ztest, DescrStatsW

# Create a DescrStatsW object with the data and standard deviation
d_salaries = DescrStatsW(salaries.values, ddof=0) #Changed: passing salaries.values instead of salaries

# Perform the z-test using the DescrStatsW object and specified value
z_stat, z_p_value = ztest(d_salaries, value=113000)  # Pass d_salaries instead of salaries and sigma

print("üîπ Z-test (bi·∫øt sigma = 15000):")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value (2 ph√≠a) = {z_p_value:.6f}")
print(f"P-value (1 ph√≠a tr√°i) = {z_p_value / 2:.6f}" if z_stat < 0 else f"P-value (1 ph√≠a ph·∫£i) = {z_p_value / 2:.6f}")
print()
```

C√¢u h·ªèi: ki·ªÉm ƒë·ªãnh gi·∫£ thi·∫øt trong tr∆∞·ªùng h·ª£p ch·ªâ bi·∫øt trung b√¨nh l∆∞∆°ng c·ªßa t·ªïng th·ªÉ 113000

ki·ªÉm ƒë·ªãnh:
H0: L∆∞∆°ng trung b√¨nh = 113000
H1: L∆∞∆°ng trung b√¨nh kh√°c 113000

```python
# Ki·ªÉm ƒë·ªãnh kh√¥ng bi·∫øt sigma (ƒë·ªô l·ªách chu·∫©n t·ªïng th·ªÉ)
# Ph·∫£i s·ª≠ d·ª•ng ph√¢n ph·ªëi t
# m·ªôt s·ªë ƒëi·ªÅu ki·ªán s·ª≠ d·ª•ng ph√¢n ph·ªëi t
# 1/ m·∫´u nh·ªè n<30
# 2/ m·∫´u l·ªõn nh∆∞ng kh√¥ng bi·∫øt sigma -> t√≠nh t s·ª≠ d·ª•ng ƒë·ªô l·ªách chu·∫©n c·ªßa m·∫´u s

# B∆∞·ªõc 1: t√≠nh z ho·∫∑c t (v·ªã tr√≠ c·ªßa trung b√¨nh m·∫´u so v·ªõi trung b√¨nh t·ªïng th·ªÉ tr√™n tr·ª•c ngang c·ªßa ph√¢n ph·ªëi)
# B∆∞·ªõc 2: t√≠nh p_value (x√°c su·∫•t H0 ƒë√∫ng, p_value c√†ng nh·ªè th√¨ H0 c√†ng d·ªÖ sai)
# B∆∞·ªõc 3: so s√°nh p_value v·ªõi alpha (m·ª©c √Ω nghƒ©a). N·∫øu p-value<alpha th√¨ b√°c b·ªè H0

# T√≠nh t
t_stat = (sample_mean - mu)/(sample_std/np.sqrt(n))
dfree = n - 1
# T√≠nh p_value d√πng ph√¢n ph·ªëi t
t_p_two_tailed = 2 * (1-stats.t.cdf(abs(t_stat),df=dfree))
print("=== Ki·ªÉm ƒë·ªãnh t (kh√¥ng bi·∫øt œÉ) ===")
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {dfree}")
print(f"P-value (2 ph√≠a): {t_p_two_tailed:.6f}")
# ƒë·∫∑t m·ª©c √Ω nghƒ©a
alpha = 0.05

if t_p_two_tailed < alpha:
  print("Lo·∫°i b·ªè H0: l∆∞∆°ng trung b√¨nh kh√°c 113000")
else:
  print("H0 ƒë√∫ng: l∆∞∆°ng trung b√¨nh = 113000")
```

ki·ªÉm ƒë·ªãnh:
H0: L∆∞∆°ng trung b√¨nh >= 113000
H1: L∆∞∆°ng trung b√¨nh < 113000

```python
t_p_right = 1 - stats.t.cdf(t_stat, df=dfree)
# ƒë·∫∑t m·ª©c √Ω nghƒ©a
alpha = 0.05

if t_p_right < alpha:
  print("Lo·∫°i b·ªè H0: l∆∞∆°ng trung b√¨nh < 113000")
else:
  print("H0 ƒë√∫ng: l∆∞∆°ng trung b√¨nh >= 113000")
```
